```
import pandas as pd
import requests
import time
import matplotlib.pyplot as plt

url = "https://uk.wikipedia.org/wiki/Населення_України"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"
}
r = requests.get(url, headers=headers)
r.encoding = "utf-8"
tables = pd.read_html(
    r.text,
    match = "Коефіцієнт народжуваності",
    decimal = ",",
    thousands = None
)
df = tables[0]

#1 Вивести перші рядки таблиці за допомогою методу head
print(f"{df.head()} \n")
time.sleep(1)

#2 Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
print(f"{df.shape} \n")
time.sleep(1)

#3 Замініть у таблиці значення "—" на значення NaN
df = df.replace("—", pd.NA)
time.sleep(1)

#4 Визначте типи всіх стовпців за допомогою dataframe.dtypes
print(f"{df.dtypes} \n")
time.sleep(1)

#5 Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
df = df.apply(pd.to_numeric, errors="ignore")
print(f"{df} \n")
time.sleep(1)

#6 Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
print(f"{df.isnull().sum() / len(df)} \n")
time.sleep(1)

#7 Видаліть з таблиці дані по всій країні, останній рядок таблиці
df = df.iloc[:-1]
print(df)
time.sleep(1)

#8 Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
df = df.fillna(df.mean(numeric_only=True))
print(f"{df} \n")
time.sleep(1)

#9 Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
print(f"{df[df["2019"] > df["2019"].mean()]["Регіон"]} \n")
time.sleep(1)

#10 У якому регіоні була найвища народжуваність у 2014 році?
print(f"{df.loc[df["2014"].idxmax(), "Регіон"]} \n")
time.sleep(1)

#11 Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
plt.bar(df["Регіон"], df["2019"])
plt.title('Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році')
plt.xticks(rotation=90)
plt.show()



def d1():
    df_line = df.set_index('Регіон').T.iloc[1:]
    df_line = df_line.apply(pd.to_numeric, errors='coerce')
    df_line.plot(kind='line', figsize=(12,6))
    plt.title('Зміна коефіцієнта народжуваності в регіонах України (1950–2019)')
    plt.xlabel('Рік')
    plt.ylabel('Коефіцієнт народжуваності (‰)')
    plt.legend(title='Регіони', bbox_to_anchor=(1.05, 1))
    plt.show()

def d2():
    df_num = df.drop(columns=['Регіон']).apply(pd.to_numeric, errors='coerce')
    mean_birth = df_num.mean(axis=0)
    mean_birth.plot(kind='bar', figsize=(12,6), color='skyblue')
    plt.title('Середній коефіцієнт народжуваності в Україні (1950–2019)')
    plt.xlabel('Рік')
    plt.ylabel('Середній коефіцієнт народжуваності (‰)')
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.show()


d1()
d2()
```
