```
import pandas as pd
import matplotlib.pyplot as plt


# Прочитайте csv файл (використовуйте функцію read_csv)
df = pd.read_csv('bestsellers with categories.csv')


# Виведіть перші п'ять рядків (використовується функція head)
print(f"{df.head()} \n")


# Виведіть розміри датасету (використовуйте атрибут shape)
print(f"{df.shape} \n")
# Відповідь:
"""датасет зберігає дані про 550 книг"""


df.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']


#4 Перевірте, чи у всіх рядків вистачає даних: виведіть кількість пропусків
#  (na) у кожному зі стовпців (використовуйте функції isna та sum)
print(f"{df.isna().sum()} \n")
# Відповідь:
""" у всіх змінних немає пропусків """


# Перевірте, які є унікальні значення в колонці genre (використовуйте функцію unique)
print(f"{df['genre'].unique()} \n")
# Відповідь:
""" 'Non Fiction' та 'Fiction' """


# Тепер подивіться на розподіл цін: побудуйте діаграму (використовуйте kind='hist')
def diagramm():
    df['price'].plot(kind='hist')
    # Визначте, яка ціна у нас максимальна, мінімальна, середня, медіанна
    # (використовуйте функції max, min, mean, median)
    print('max =', df['price'].max(),
          '\nmin =', df['price'].min(),
          '\nmean =', df['price'].mean(),
          '\nmedian =', df['price'].median())
    print("\n")
    plt.show()
diagramm()
# Відповідь:
""" max = 105
    min = 0
    mean = 13.1
    median = 11.0 """


print(df['user_rating'].max())
print(df[df['user_rating']==df['user_rating'].max()].shape[0])
print(df.loc[df['reviews'].idxmax(), 'name'])
print(df.loc[df[df['year']==2015]['price'].idxmax(), 'name'])
print(df[(df['year']==2010) & (df['genre']=='Fiction')].shape[0])
print(df[(df['user_rating']==4.9) & (df['year'].isin([2010,2011]))].shape[0])
print(df[(df['year']==2015) & (df['price']<8)].sort_values('price').iloc[-1]['name'], "\n")
# Відповідь:
""" найвищий рейтинг 4.9
    такий рейтинг мают 52 книг
    Where the Crawdads Sing має найбільше відгуків
    Publication Manual of the American Psychological Association, 6th Edition найдорожча книга 2015 року
    20 книг жанру Fiction потрапили до Топ-50 у 2010 році
    одна книга з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках

    Old School (Diary of a Wimpy Kid #10) остання у відсортованому списку """



price_stats = df.groupby('genre')['price'].agg(['min', 'max'])
print( price_stats.loc['Fiction', 'max'])
print(price_stats.loc['Fiction', 'min'])
print(price_stats.loc['Non Fiction', 'max'])
print(price_stats.loc['Non Fiction', 'min'], "\n")

books_per_author = df.groupby('author')['name'].agg(count='count')
print(books_per_author.shape)
print(books_per_author['count'].idxmax())
print(books_per_author['count'].max(), "\n")

rating_per_author = df.groupby('author')['user_rating'].agg(mean_rating='mean')
print(rating_per_author['mean_rating'].idxmin())
print(rating_per_author['mean_rating'].min(), '\n')

final_df = pd.concat([books_per_author, rating_per_author], axis=1)
final_df_sorted = final_df.sort_values(by=['count', 'mean_rating'], ascending=[True, True])
print(final_df_sorted.index[0])

# Відповідь:
""" Максимальна ціна (Fiction): 82
    Мінімальна ціна (Fiction): 0
    Максимальна ціна (Non Fiction): 105
    Мінімальна ціна (Non Fiction): 0

    Розмірність таблиці (рядки, стовпці): (248, 1)
    Автор з найбільшою кількістю книг: Jeff Kinney
    Кількість книг цього автора: 12

    Автор з мінімальним середнім рейтингом: Donna Tartt
    Мінімальний середній рейтинг: 3.9

    Перший автор у відсортованому списку: Muriel Barbery """


def countgeneres():
    df['genre'].value_counts().plot(kind='bar')
    plt.title('Кількість книг за жанрами')
    plt.xlabel('Жанр')
    plt.ylabel('Кількість книг')
    plt.show()

def midrating():
    df.groupby('year')['user_rating'].mean().plot(kind='line')
    plt.title('Середній рейтинг книг по роках')
    plt.xlabel('Рік')
    plt.ylabel('Середній рейтинг')
    plt.show()

def deprevonprice():
    df.plot(kind='scatter', x='price', y='reviews')
    plt.title('Залежність кількості відгуків від ціни')
    plt.xlabel('Ціна')
    plt.ylabel('Кількість відгуків')
    plt.show()

countgeneres()
midrating()
deprevonprice()





```
